# the most common student name 

```python
m_std = open('male_students.txt', encoding='utf-8').read().strip().split('\n')
print('there are {} names'.format(len(m_std)))
m_dict = {}
for name in m_std:
    if name in m_dict:
        m_dict[name] += 1 
    else:
        m_dict[name] = 1 
m_dict
```

# write a program to let the user to enter a text, the coun the word frequency for the entered text 

```python
text = input('please enter a text: ')
words = text.split()
count_dict = {}
for word in words:
    count_dict[word] = count_dict.get(word, 0) + 1 
print(count_dict)
```

```python
count_dict.items()
```

```python
x, y = (2, 3)
```

```python
x
```

```python
y
```

```python
for k, v in count_dict.items():
    print('word: {}\t frequency: {}'.format(k, v))
```

```python
print('word \t frequency')
print('----------------')
for word, freq in count_dict.items():
    print('{} \t {}'.format(word, freq))
```
